```python
import numpy as np
import matplotlib as plt
import pandas as pd
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,GRU
import matplotlib.pyplot as plt 
```

```python
#Cargar datos
company = 'AMZN'
ticker = yf.Ticker(company) #yf nos permite obtener los datos de amazon
hist = ticker.history(start = '2012-1-1', end='2020-1-1')
hist
```

```python
#Preparar datos (valores normalizados entre cero y uno)
#Se preparan para que puedan ser leidos por la red neuronal
#Usamos MinMaxScaler para ajustar los valores al rango de 0 a 1
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(hist['Close'].values.reshape(-1,1))
scaled_data
#Con Fit transfors tranformamos los valores de los precios de cierra
#reshape nos ayuda a poner los objetos en un array
```

```python
#construir el modelo
#almacenamos los días con los que queremos entrenar la red neuronal
prediction_days = 60

#Estas variables permiten entrenar la red
x_train = []
y_train = []

#iniciamos un bucle que comienza en 60 y va hasta la longitud del data
#x_train array formado por los precios de los ultimos 60 dias, en la columna cero están los ceros
#y_train es el precio del dia 60
for x in range(prediction_days, len(scaled_data)):
    x_train.append(scaled_data[x-prediction_days:x,0])
    y_train.append(scaled_data[x,0])

#el x train es una lista no un array por lo que lo convertimos en un array
#transformamos las dimensiones del array para que sea de tres dimensiones debido a que así lo pide la red neuronal
x_train, y_train = np.array(x_train),np.array(y_train)
x_train = np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))

x_train.shape
#(1952,60,1) 1952 filas y en cada fila hay 60 valores


#contruimos el modelo
#iniciamos un modelo de tipo secuencial
model = Sequential()
#agregamos capas, la primera es una capa de tipo GRU con 50 unidades
#return sequences = true permite a la red neuronal comunicarse hacia atás y hacia adelante
model.add(GRU(units=50,return_sequences = True, input_shape=(x_train.shape[1],1)))
#agregamos una capa de tipo dropout
model.add(Dropout(0.2))
model.add(GRU(units=50,return_sequences = True))
model.add(Dropout(0.2))
model.add(GRU(units=50))
model.add(Dropout(0.1))
model.add(Dense(units=1))#capa que nos predice el valor final

#compilamos el modelo
model.compile(optimizer='adam', loss='mean_squared_error')
#llamamos el modelo para entrenarlo a 25 epocas
model.fit(x_train,y_train,epochs=25,batch_size=32)
```

```python
#vamos a probar la eficiencia de la red

#cargar los datos del test
hist_test = ticker.history(start = '2020-1-1', end='2022-8-21')
actual_prices = hist_test["Close"].values

#concatenamos los valores de los años anteriores con los añoes e prueba
total_dataset = pd.concat((hist['Close'],hist_test['Close']),axis=0)
#los transformamos para que sean leidos por la red
model_inputs = total_dataset[len(total_dataset)-len(hist_test)-prediction_days:].values
model_inputs = scaler.transform(model_inputs.reshape(-1,1))


```

```python

#realizamos predicciones 
x_test=[]

for x in range(prediction_days,len(model_inputs)):
    x_test.append(model_inputs[x-prediction_days:x,0])

x_test = np.array(x_test)
x_test = np.reshape(x_test,(x_test.shape[0],x_test.shape[1],1))

#creamos la variables para almacenar los precios predecidos
predicted_prices = model.predict(x_test)
#transformamos inversamente los datos 
predicted_prices = scaler.inverse_transform(predicted_prices)
predicted_prices
```

```python
#comparamos los precios reales con los predecidos por la red neuronal
plt.plot(actual_prices,color="black",label=f"{company} real prices")
plt.plot(predicted_prices,color="blue", label=f"{company} predicted prices")
plt.show()
```
